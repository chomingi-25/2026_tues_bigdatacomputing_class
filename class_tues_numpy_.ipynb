# 1. 넘파일 다차원 배열(ndarray) 객체 생성
- import numpy as np
- 리스트을 이용하여 다차원 배열 생성
- np.array(리스트)
  - 리스트 → 배열로 바꾸고 싶을 때
  - 단순 저장 → 수학 연산이 필요할 때
  - 데이터 분석/머신러닝 모델에 데이터를 입력할 때
- 속성 :
  - ndim : 배열 축 혹은 차원의 개수
  - shape : 배열의 형상을 (n, m) 형식의 튜플형으로 표시
  - size :
    - 배열 원소의 개수
    - (m,n) 형상 배열의 size = m*n
  - dtype : 배열 원소의 자료형을 기술
  - astype : 배열 원소 형 변환

```python
# 파이썬의 리스트 vs 넘파이 다차원 배열
import numpy as np
# 1. 데이터 생성
list_a=[1, 2, 3]
print(list_a)

arr_a=np.array(list_a) # 1차원 넘파이 배열(벡터) 생성
print(arr_a)
print(arr_a.ndim) # 1
print(arr_a.size) # 3
print(arr_a.shape) #(3,)
print(arr_a.dtype) # int64
print(type(arr_a))
# 2. 연산 * 처리
print(list_a*2) # 단순 복제
print(arr_a*2) # 원소별 곱셈
```

```python
# 다차원 구조 리스트와 넘파이 2차원 배열
import numpy as np
# 중첩된 리스트 생성 : (2*3 구조: 2행 3열)
matrix_list=[ [1, 2, 3], [4, 5, 6] ]
print(matrix_list)
print("="*100)
#넘파이 2차원 배열 생성
matrix=np.array(matrix_list)
print(matrix)
print(matrix.ndim)
print(matrix.shape)
print(matrix.size)
print(matrix.dtype)
```

```python
# 데이터의 타입을 강제로 변환
# 정수 리스트 -> 실수형으로 생성
import numpy as np
arr=np.array([1, 2, 3], dtype='float32')
print(arr)
```

# 2. 넘파이 다차원 배열 간 사칙 연산
  - 피연산자 다차원 배열들의 형상이 같다는 가정
  - +, -, *, / 연산 수행
  - 배열 프로그래밍
  - 선형대수의 벡터화 연산
  - 벡터을 병렬 연산 수행
    - 데이터 병렬 처리

```python
import numpy as np
# 두 개의 리스트을 생성
store_a=[20, 30, 15]
store_b=[70, 90, 70]

# 리스트을 이용하여 넘파이 1차원 배열 생성
np_store_a = np.array(store_a)
np_store_b=np.array(store_b)
print(np_store_a)
print(np_store_b)
print("="*100)

# + 연산
print(np_store_a + np_store_b)
print("="*100)
# - 연산
print(np_store_a-np_store_b)
print("="*100)
# * 연산
print(np_store_a*np_store_b)
print("="*100)
# / 연산
div=np_store_a/np_store_b
print(div.round(2))
```

# 6. 다차원 배열의 지정된 축 방향으로 원소들을 뒤집기
  - np.flip(배열객체, axis=)

```python
import numpy as np

# (4, 5) 이미지 데이터 생성
image=np.arange(1, 21).reshape((4, 5))
print(image)

# 좌우 반전
h_flip=np.flip(image, axis=1)
print(h_flip)

# 상하 반전
v_flip=np.flip(image, axis=0)
print(v_flip)

# 전체 반전 (상하좌우 모두)
all_flip=np.flip(image)
print(all_flip)
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

# 15. 넘파일 배열의 결합
- np.concatenate(튜플)
- np.vstack(튜플 또는 리스트)
- np.hstack(튜플 또는 리스트)
- np.r_[ 콤마 사용하여 나열 ]
- np.c_[콤마 사용하여 나열]
- np.column_stack(리스트)


```python
import numpy as np
#----------------------
# Ex1. 1차원 배열
#----------------------

a=np.array([1, 2, 3])
b=np.array([4, 5, 6])

# (1) 세로로 쌓기 -> 수직방향 행 추가
res_vstack=np.vstack((a, b))
print(res_vstack)

# 가로로 쌓기 -> 수평방향으로 열 추가
res_col=np.hstack((a, b)) # (6,)
print(res_col)
res_hstack=np.column_stack((a, b))
print(res_hstack) # (3, 2)

#---------------------
# Ex1. 형태가 다른 배열 결합
#---------------------
arr_1=np.array([1, 2, 3]) #(3,)
arr_2=np.array([[4, 5, 6], [7, 8, 9]]) #(2, 3)

# 방법 1
# (1) arr_1 -> (1, 3) 형태로 변환
arr_1_2d=arr_1.reshape(1, 3)

# (2) arr_2 & arr_1_2d 결합 (axis=0) => (3, 3)
combined = np.concatenate((arr_1_2d, arr_2), axis=0)
print(combined)

# 방법2
v_combined=np.vstack((arr_1, arr_2))
print(v_combined)
```

## 프로그래밍 문제 : 시계열 데이터 기반 봇 탐지 데이터 셋 구축 프로그래밍
-설명: 웹사이트 접속 로그를 분석하여 자동화된 봇과 실제 사람을 구분하는 기계학습 모델링하는데 필요한 데이터 셋을 구축 문제
-각 사용자로부터 수집한 데이터는 100개의 시점에서 이루어진 로그 수치 데이터
-하기 조건에 맞는 데이터 전처리 함수을 작성하시오.
- 봇로그와 사람 로그을 비교하는 히스트그램으로 시각화하시오.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.font_manager as fm
import pandas as pd

def setup_korean_font():
    """한글 폰트 설정 (수정 버전)"""
    font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

    # 1. 폰트 파일 존재 여부 확인 및 설치
    if not os.path.exists(font_path):
        print("Installing Nanum fonts...")
        os.system('sudo apt-get install -y fonts-nanum > /dev/null')

    # 2. 폰트 매니저에 폰트 추가 (문제의 원인 해결)
    # FontProperties 대신 font_manager.fontManager.addfont를 사용.
    fm.fontManager.addfont(font_path)

    # 3. 전역 설정 적용
    # 'NanumBarunGothic'은 나눔폰트 파일의 실제 이름입니다.
    plt.rc('font', family='NanumBarunGothic')
    plt.rcParams['axes.unicode_minus'] = False
    print("✅ 한글 폰트 설정 완료 (Matplotlib 연동 성공)")
```

```python
"""
- 데이터 수집 & 구축:
(1) bot_1(50 개), bot_2(30개) 샘플을 생성하고 이를 하나로 합치세요. (평균 0, 표준편차 0.001인 정규분포)
(2) human(80개) 샘플을 생성하세요. (평균 0, 표준편차 0.05인 정규분표)
(3) 레이블링: 봇 데이터는 1, 사람  데이터는 0으로 레이블을 지정하세요.
(4) 데이터 통합: 모든 샘플(160개)과 레이블을 하나의 2차원 배열로 결합하세요
(5) 데이터 셔플링 (추가): 모델 학습 시 편향을 방지하기 위해 데이터를 무작위로 섞으세요.
"""
def build_bot_detection_dataset():
  #[1] 데이터 생성
  np.random.seed(42)

  # 시계열 데이터 봇 데이터
  bot_1=np.random.normal(0, 0.001, (50, 100))
  bot_2=np.random.normal(0, 0.001, (30, 100))
  all_bots=np.vstack((bot_1, bot_2)) # (80, 100)

  # 시계열 사람 데이터
  human=np.random.normal(0, 0.05, (80, 100))

  # [2]. 넘파이의 리덕션 집계
  # 모든 데이터셋을 합치기
  X_data = np.vstack((all_bots, human)) # (160, 100)
  # 합쳐진 2차원 배열의 통계 분포 : 평균, 표준편차
  row_means=np.mean(X_data, axis=1).reshape(-1, 1) # (160, 1)
  row_stds=np.std(X_data, axis=1).reshape(-1, 1) # (160, 1)

  # [3] 레이블 생성 (봇 = 1, 사람 = 0)
  y_data=np.concatenate([np.ones(80), np.zeros(80)]).reshape(-1, 1) #(160, 1)

  # [4] 최종 결합 : 데이터 (100) + 평균(1) + 표준편차(1) + 레이블(1) => (160, 103)
  dataset=np.column_stack(((X_data, row_means, row_stds, y_data)))

  #  [5] 데이터 셔플링
  np.random.shuffle(dataset)
  print(dataset.shape) # (160, 103)
  print(dataset[:5, 100: ])
  print("✅ Numpy 리덕션 및 데이터셋 구축 완료:")

  return dataset
```

```python
# 시각화 로직
def visualize_by_numpy(dataset):
  """ Numpy dataset을 활용한 봇/사람 데이터 분포 시각화"""
  # 표준편차을 이용한 시각화
  std_vals=dataset[:, -2].astype(float) # 표준편차 데이터 추출 및 형변환
  labels=dataset[:, -1].astype(float) # 레이블 데이터 추출 및 형변환

  # 불리언 인덱싱을 사용하여 그룹 이진으로 분리
  bot_stds = std_vals[labels == 1]
  human_stds=std_vals[labels==0]

  # 시각화 설정
  plt.figure(figsize=(10, 6))

  # 히스트그램 그리기
  plt.hist(bot_stds, bins=20, alpha=0.7, label='봇', color='red', edgecolor='black')
  plt.hist(human_stds, bins=30, alpha=0.5, label='사람', color='blue', edgecolor='black')
  plt.xlabel('표준편차')
  plt.ylabel('사용자수')
  plt.grid(True)
  plt.legend()
  plt.tight_layout()
  plt.show()
  print("✅ 시각화 완료")
```

```python
#--- 실행 파트 -----
if __name__=="__main__":
  # 환경 설정
  setup_korean_font()
  # 데이터 셋 구축 (Numpy 2차원 배열 (160, 103))
  numpy_dataset=build_bot_detection_dataset()
  # Numpy 2차원 배열 기반 시각화
  visualize_by_numpy(numpy_dataset)
```
